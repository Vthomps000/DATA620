```python
import sys
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
import copy
import itertools
import pylab as pyt
```

### Creating the edge array.

Listing all edges between nodes by visualizing the relationships.

```python
edgesarray = ['1 2','1 5','1 6',
         '2 3','2 4','2 5','2 6','2 7',
         '3 4','3 7', 
         '4 5','4 7','4 8',
         '5 6','5 8',
         '6 7',
         '8 9',
         '9 10']
```

### Labeling the edges.

Each node is set and defined by a label (name).

```python
edgelabel={}
edgelabel[1] = "Ed"
edgelabel[2] = "Diane"
edgelabel[3] = "Carol"
edgelabel[4] = "Fernando"
edgelabel[5] = "Garth"
edgelabel[6] = "Beverly"
edgelabel[7] = "Andre"
edgelabel[8] = "Heather"
edgelabel[9] = "Ike"
edgelabel[10] = "Jane"
```

### Forming the network 

The edges are parsed into a network.

```python
names = nx.parse_edgelist(edgesarray,nodetype = int)
```

### Mapping

Using the cartesian plane, a map is created by fixing positions in the plane.

```python
fixed_positions = {1:(1,1),
                   2:(1,3),
                   3:(1,5),
                   4:(2,4),
                   5:(2,2),
                   6:(0,2),
                   7:(0,4),
                   8:(3,3),
                   9:(4,3),
                   10:(5,3)}
fixed_nodes = fixed_positions.keys()
pos = nx.spring_layout(names,pos=fixed_positions, fixed = fixed_nodes)
```

### Plot

```python
import warnings
from matplotlib.cbook.deprecation import MatplotlibDeprecationWarning

warnings.filterwarnings("ignore", category=MatplotlibDeprecationWarning)

fig = plt.figure(figsize=(20,20))
ax = plt.subplot(111,aspect = 'equal')
plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0)

nx.draw(names, pos, labels=edgelabel, 
        with_labels = True, 
        node_size=15000, 
        node_color="white", 
        edgecolors="black", 
        font_size=20,
        width=10)

plt.show()
```
